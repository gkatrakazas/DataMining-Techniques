```python
def euclidean_distance(p1, p2):
    return np.sqrt(np.sum((p1 - p2)**2))
```

```python
def dtw_distance(s1, s2):
    M, N = len(s1), len(s2)
    D = np.zeros((M, N))
    for i in range(M):
        for j in range(N):
            D[i, j] = euclidean_distance(s1[i], s2[j])
    
    DTW = np.zeros((M, N))
    DTW[0, 0] = D[0, 0]
    for i in range(1, M):
        DTW[i, 0] = DTW[i-1, 0] + D[i, 0]
    for j in range(1, N):
        DTW[0, j] = DTW[0, j-1] + D[0, j]
    for i in range(1, M):
        for j in range(1, N):
            DTW[i, j] = D[i, j] + min(DTW[i-1, j], DTW[i, j-1], DTW[i-1, j-1])
            
    return DTW[-1, -1]
```

```python
result = []

print('Started time')
startTime = time.time()

for i, row in dtw_test.iterrows():
    if i%100==0:
        print(i)
    id,s1, s2 =row['id'], row['series_a'], row['series_b']
    s1 = np.array(ast.literal_eval(s1))
    s2 = np.array(ast.literal_eval(s2))
    result.append([id, dtw_distance(s1, s2)])
    
total_time = time.time() - startTime
print('Total time took {} seconds'.format(total_time))

result_df = pd.DataFrame(result, columns=['id', 'distance'])
result_df.to_csv('/content/drive/My Drive/DataAnalytics/Output/Output_3/dtw.csv', index=False)
```
